```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gr´afica de l´ınea simple")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 4, 2, 5]
plt.scatter(x, y)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gr´afica de dispersi´on")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Generar valores de x
x = np.linspace(0, 10, 100)
# Definir funciones
y1 = x**2
y2 = 2*x + 1
# Crear figura
plt.figure()
# Gr´afica de l´ınea
plt.plot(x, y1, label="y = x^2")
# Gr´afica de dispersi´on
plt.scatter(x, y2, color="Green",label="y = 2x + 1")
# Etiquetas y t´ıtulo
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gr´aficas combinadas")
plt.legend()
plt.grid()
plt.show()
```

```python
# Generar valores de x
x1 = np.linspace(0, 10, 100)
# Definir funciones
y3 = np.power(x1, 2)
y4 = np.sin(x1)
y5 = np.exp(-x1)
# Crear figura
plt.figure()
# Gr´afica de l´ınea
plt.plot(x1, y3, y4, label="y = x^2")
# Gr´afica de dispersi´on
plt.scatter(x1, y4, color="Green",label="y = 2x + 1")
# Etiquetas y t´ıtulo
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gr´aficas combinadas")
plt.legend()
plt.grid()
plt.show()
```
