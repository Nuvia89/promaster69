```python
## import numpy as np

def funcionmatriz(matri):
    # Obtener dimensiones usando .shape
    num_filas, num_cols = matri.shape

    # Extraer columnas con slicing
    # De columna a lista
    ListaColumnas = [matri[:, i] for i in range(num_cols)]

    return ListaColumnas, num_cols, num_filas

try:
    i = input("Número del dataset a leer: ")
    file_name = f"dataset{i}.txt"

    datos = np.loadtxt(file_name)
    cols, total_cols, total_rows = funcionmatriz(datos)

    print(f"Reporte")
    print(f"Número total de columnas: {total_cols}")
    print(f"Número de elementos por columna: {total_rows}")

    for i, col in enumerate(cols):
        print(f"Columna {i} (Arreglo 1D): {col}")

except FileNotFoundError:
    print(f"Error: El archivo 'dataset{i}.txt' no fue encontrado.")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```
