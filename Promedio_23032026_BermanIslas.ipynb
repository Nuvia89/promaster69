```python
import numpy as np

def calcular_promedio_numpy():
    try:
        n = int(input("¿Cuántos números deseas promediar? "))
        
        if n <= 0:
            print("Por favor, ingresa un número mayor a cero.")
            return

        numeros = []
        for i in range(n):
            numero = float(input(f"Ingresa el número {i + 1}: "))
            numeros.append(numero)

        arreglo_numpy = np.array(numeros)
        
        promedio = np.mean(arreglo_numpy)
        
        print(f"\nEl promedio de los {n} números es: {promedio}")

    except ValueError:
        print("Error: Por favor, ingresa únicamente valores numéricos.")

calcular_promedio_numpy()
```
