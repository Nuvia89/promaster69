```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

data1 = np.loadtxt('dataset1.txt')
data2 = np.loadtxt('dataset2.txt')
c1=data1[:,0]
d1=data2[:,0]
c1=c1+5
c1=2*c1
producto=c1*d1
f=c1[1:5]
len(f)

dimension=data1.shape
n_elementos = data1.size
print(dimension)
print(type(dimension))

filas = dimension[0]
columnas = dimension[1]
print(filas)
print(columnas)

data1[0][0] = 3.4
lista1 = [4,5,6,7.6,7,9]
lista1[3] = 10
lista2x2=[[2,3,4],[6,7,8]]
t2x2=([2,3,4],[6,7,8])
t2c2=((2,3,4),(6,7,8))
print(type(lista2x2))
print(type(t2x2))
print(type(t2c2))
print(t2x2[0][0])
try:
    t2x2[0][0]=10
except TypeError as e:
    print(f"Error esperado en tupla: {e}")

print(t2x2[0][0])
print(type(data1))

with open('dataset1.txt', 'r') as f1, open('dataset2.txt', 'r') as f2:
    print(type(f1))
    data1 = np.loadtxt(f1)
    data2 = np.loadtxt(f2)
    print(type(data1))

print(data1.shape)
print(data1.size)

c1 = data1[:,0]
c2 = data1[:,1]
c3 = data1[:,2]
c4 = data1[:,3]

d1 = data2[:,0]
d2 = data2[:,1]
d3 = data2[:,2]
d4 = data2[:,3]

print(len(c1))
print(c1.shape)

plt.figure(1)
plt.plot(c1)
plt.title("Columna 1 Dataset 1")
plt.figure(2)
plt.plot(c2)
plt.title("Columna 2 Dataset 1")
plt.figure(3)
plt.plot(c3)
plt.title("Columna 3 Dataset 1")
plt.figure(4)
plt.plot(c4)
plt.title("Columna 4 Dataset 1")

def graficar_6_tipos(data, titulo_base):
    col1, col2, col3, col4 = data[:,0], data[:,1], data[:,2], data[:,3]
    
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle(f'Análisis de 4 tipos: {titulo_base}', fontsize=16)

    axs[0, 0].scatter(col1, col2, alpha=0.5, color='blue', s=10)
    axs[0, 0].set_title('1. Dispersión (C1 vs C2)')

    axs[0, 1].hist(col2, bins=30, color='green', edgecolor='black')
    axs[0, 1].set_title('2. Histograma (C2)')

    axs[1, 0].step(range(len(col3[:100])), col3[:100], color='red')
    axs[1, 0].set_title('3. Pasos (C3 - Primeros 100)')

    axs[1, 1].boxplot([col1, col2, col3, col4])
    axs[1, 1].set_xticklabels(['C1', 'C2', 'C3', 'C4'])
    axs[1, 1].set_title('4. Boxplot (Todas)')


    plt.tight_layout(rect=[0, 0.03, 1, 0.95])


graficar_6_tipos(data2, "Dataset 2")

plt.show()
```
